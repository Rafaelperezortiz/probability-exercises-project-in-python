## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np

import random

contador = 0 #crear contador

for i in range(1000):   #simulacion 1000 veces
    dado1 = random.randint(1, 6) #dado1 con las posibilidades
    dado2 = random.randint(1, 6) #dado2 con las posibilidades
    suma = dado1 + dado2   #suma de los resultados de las posibilidades

    if (suma > 7) or (suma % 2) == 0:#comprobacion de si es mayor que 7 o numero par
        contador +=1

probabilidad = contador / 1000 

print("La probabilidad estimada de obtener una suma mayor a 7 o un numero par es de" ,probabilidad,)
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

#print(ball_box)
            
# sacar una bola al azar
import random
escenario_1 = 0
escenario_2 = 0
random.seed(17)
for i in range(1000):
    random_balls = random.choices(list(ball_box.values()),k = 5)
    white_count = random_balls.count("White")
    red_count = random_balls.count("Red")
    green_count = random_balls.count("Green")
    if (white_count == 3) and (red_count == 2):
        escenario_1 += 1
    elif (white_count == 5) or (red_count == 5) or (green_count == 5):
        escenario_2 += 1

print("la probabilidad de que salgan 3 blancas y 2 rojas es de",(escenario_1/1000)*100,"%")
print("la probabilidad de que salgan las 5 bolas del mismo color es de ",(escenario_2/1000)*100,"%")        



```
